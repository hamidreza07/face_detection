```python
from facetorch import FaceAnalyzer
from omegaconf import OmegaConf
from torch.nn.functional import cosine_similarity
from typing import Dict
import operator
import torchvision
```

```python

path_img_output="/test_output.jpg"
path_config="gpu.merged.config.yaml"


cfg = OmegaConf.load(path_config)
```

```python
import os

analyzer = FaceAnalyzer(cfg.analyzer)
path = ['brigh_cont','brigh_cont - rotate','brigh_gray_rotate','gray_scale','gray_scale - rotation','main_rotation']
for file in path:
    for image in os.listdir(file):
        path_img_input=f"{file}/{image}"
            # initialize

        # warmup
        response = analyzer.run(
                path_image=path_img_input,
                batch_size=10,
                fix_img_size=True,
                return_img_data=True,
                include_tensors=True,
                path_output=path_img_output,
        )
        pil_image = torchvision.transforms.functional.to_pil_image(response.img)
        os.makedirs(f"res_{file}",exist_ok=True)
        pil_image.save(f'res_{file}/{image}')
    
```
